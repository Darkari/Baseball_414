```r
base.dat<-read.csv('baseball.csv',header = TRUE)
wea.dat<-read.csv('weather.csv',header = TRUE)
base.dat<-merge(base.dat,wea.dat,by.x = 'event_date',by.y = 'date')
base.dat$event_date<-as.Date(base.dat$event_date)
n<-nrow(base.dat)
base.dat$timeofday <- ifelse(as.numeric(substr(base.dat$event_time,1,2))<19,c("Afternoon"),c("Evening"))
base.dat$weekday = weekdays(base.dat$event_date)
baseball<-base.dat
baseball$payrollratio <- baseball$opponent_payroll / baseball$braves_payroll

```

```r
#question 2 predicting win/loss

#logit model comparsion
#we try different combination of predictors and compare the feedback from the model

# Model 1
modb1 <- glm(formula = win ~ opponent_payroll,
             family = binomial(link = "logit"), data = subset(baseball, 
                                                              season <= 2014))
summary(modb1)
```

```r

# Model 2
modb2 <- glm(formula = win ~ opponent_payroll+braves_payroll+opponent,
             family = binomial(link = "logit"), data = subset(baseball,
                                                              season <= 2014))
summary(modb2)

```

```r
# Model 3 
#we try some dummy varables for some particular opponents
oppo.dat<-baseball
for(level in unique(oppo.dat$opponent)){
  oppo.dat[paste("dummy", level, sep = "_")] <- ifelse(oppo.dat$opponent == level, 1, 0)
}
modc3<-glm(win~`dummy_Los Angeles Dodgers`+`dummy_Philadelphia Phillies`+ `dummy_San Francisco Giants`+`dummy_Los Angeles Dodgers`+opponent_payroll,data = oppo.dat,family = binomial())

summary(modc3)
```

```r
# Model 4
modb4 <- glm(formula = win ~ opponent_payroll+braves_payroll+opponent_rank,
             family = binomial(link = "logit"), data = subset(baseball,
                                                              season <= 2014))
summary(modb4)
```

```r
# Model 5
modb5 <- glm(formula = win ~ opponent_payroll+braves_payroll+opponent_rank+braves_rank,
             family = binomial(link = "logit"), data = subset(baseball,
                                                              season <= 2014))
summary(modb5)
```

```r
# Model 6
modb6 <- glm(formula = win ~ opponent_payroll+braves_payroll+opponent_rank+timeofday,
             family = binomial(link = "logit"), data = subset(baseball,
                                                              season <= 2014))
summary(modb6)
```

```r
# Model 7
modb7 <- glm(formula = win ~ opponent_payroll+braves_payroll+opponent_rank+payrollratio,
             family = binomial(link = "logit"), data = subset(baseball,
                                                              season <= 2014))
summary(modb7)
```

```r
#predict win with 2015-2016 data
pred_win=as.numeric(predict(modb1, newdata = baseball[baseball$season%in%c(2015,2016),], type="response")>0.5)
pred_win
```

```r
#actual win data
actual_win=baseball$win[baseball$season%in%c(2015,2016)]
actual_win
```

```r

#comparsion table
table(pred_win,actual_win)
```

```r
#bestglm part (details can be seen in https://cran.r-project.org/web/packages/bestglm/bestglm.pdf)
#this package can be used to find the best glm model in terms of Information Criterion by Cross Validation
#the search of model is using leaps function so bestglm is similiar with regsubset in 'leaps' package
#data need to be rearrange and some catorgical varable are deleted
install.packages('bestglm')
library(bestglm)
```

```r
#factorize character data
baseball$weekday<-as.factor(baseball$weekday)
baseball$timeofday<-as.factor(baseball$timeofday)
```

```r
#manually create dummy variables for catagorical variables to fit the data input requirement
#opponent dummy variables are not included since it will take so much time for processing about 30 more dummy variables.
#We have also try including those opponent dummy variables and they proved to be insignificant so we do not include opponent in our coding

for(level in unique(baseball$weekday)){
  baseball[paste("dummy", level, sep = "_")] <- ifelse(baseball$weekday == level, 1, 0)
}
for(level in unique(baseball$weather)){
  baseball[paste("dummy", level, sep = "_")] <- ifelse(baseball$weather == level, 1, 0)
}
for(level in unique(baseball$timeofday)){
  baseball[paste("dummy", level, sep = "_")] <- ifelse(baseball$timeofday == level, 1, 0)
}
```

```r
baseball$winn<-baseball$win

#spliting training and testing dataset
#Some varables are filtered because of insignificance or corecion with existing variables

dat.tr<-subset(baseball,event_date>="2011-01-01"& event_date<="2014-12-31",select = -c(event_time,event_id,event_date,season,win,opponent,weekday,timeofday,weather,dummy_Sunny,dummy_Afternoon,dummy_Friday))
dat.te<-subset(baseball,event_date>="2015-01-01",select = -c(event_time,event_id,event_date,season,win,opponent,weekday,timeofday,weather,dummy_Sunny,dummy_Afternoon,dummy_Friday))
```

```r
a<-bestglm(dat.tr, IC='AIC')
a
```
